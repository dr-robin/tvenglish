```python
import time

def sub2srt(file):
    timestamp=""
    sub=""
    count=0
    #with open(file, 'rb') as f:
     #   rb = f.read()
      #  encoding = chardet.detect(rb)['encoding']
        
    with open(file, 'r', encoding="ascii", errors="surrogateescape") as f:
        data = f.readlines()
        
    for i in data:
        pattern =r'(?!\{).\d*(?=\})'
        match = re.findall(pattern, i)
        if match:
            start = match[0]
            end = match[1]
            start = int(start)/30
            end = int(end)/30
            start = time.strftime('%H:%M:%S', time.gmtime(start))
            end = time.strftime('%H:%M:%S', time.gmtime(end))
            timestamp = start + ' --> ' + end
        sub = i.split('}')[-1]
        if '|' in sub:
            sub = sub.replace('|','')
        count += 1
        
        fname = file.split('.')[0]
        fname = fname + '.srt'
        with open(fname, 'a', encoding="ascii", errors="surrogateescape") as tmp:
            tmp.write("%s\n%s\n%s\n" % (str(count), timestamp, sub))
    print(f"{file} converted to {fname}")
```
